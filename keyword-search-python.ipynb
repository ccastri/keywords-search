```python


# Define the keywords and timeframes as a list of dictionaries
keyword_timeframes = [
    {'keyword': 'Abogado', 'timeframe': '2022-09-04 2022-09-10'},
    {'keyword': 'Pension', 'timeframe': '2022-09-18 2022-09-24'},
    {'keyword': 'Salud', 'timeframe': '2022-09-18 2022-09-24'}
]

# Build the payload using the list of dictionaries
pytrends.build_payload(keyword_timeframes)

# Fetch multi-range interest over time data
multirange_data = pytrends.multirange_interest_over_time()

# Print or work with the multi-range data
print(multirange_data)
```
